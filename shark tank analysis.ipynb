```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


def remove_outliers(data, column):
    Q1 = data[column].quantile(0.25)  
    Q3 = data[column].quantile(0.75)  
    IQR = Q3 - Q1 
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

filtered_df = df.copy()


for column in df.select_dtypes(include=['float64', 'int64']).columns:
    
    filtered_df = remove_outliers(filtered_df, column)
    plt.figure(figsize=(8, 4))  
    sns.boxplot(data=filtered_df[[column]], width=0.6, palette="Set2")  
    plt.title(f"Box Plot for {column} (Outliers Removed)", fontsize=16)  
    plt.xlabel("Values", fontsize=12) 
    plt.ylabel(column, fontsize=12)  
    plt.xticks(fontsize=10)  
    plt.yticks(fontsize=10) 
    plt.tight_layout()
    plt.show()  
```

# Analysis of Rejected Pitches

# 1.  What percentage of pitches were rejected outright?

```python
total_pitches = len(df)
rejected_pitches = df[df["Deal"] == 0]

# Percentage of pitches rejected outright
rejected_percentage = round(((len(rejected_pitches) / total_pitches) * 100),2)
print(f'{rejected_percentage}% pitches rejected outright')
```

# 2. Which product categories had the highest rejection rates?

```python
category_rejection_counts = rejected_pitches['Idea'].value_counts()
category_rejection_rates = (category_rejection_counts / len(rejected_pitches) * 100).round(2)
sorted_rejection_rates = category_rejection_rates.sort_values(ascending=False)
print("Product Categories with Highest Rejection Rates (Based on Rejected Pitches Data):")
print(sorted_rejection_rates.head())
```

# 3. What is the average investment and equity percentage for rejected pitches?


```python
avg_investment_rejected = rejected_pitches["Ask_Investment_Amount"].mean()
avg_equity_rejected = rejected_pitches["Ask_Equity"].mean()
print(round(avg_investment_rejected,2))
print(round(avg_equity_rejected,2))
```

# 4. Are there patterns in the financial asks (debt, equity, investment) for rejected pitches?

```python
correlation_matrix = rejected_pitches[['Ask_Investment_Amount','Ask_Equity','Ask_Valuation']].corr()
print("Summary of the rejected pitches")
print(rejected_pitches[['Ask_Investment_Amount','Ask_Equity','Ask_Valuation']].describe())
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

# 3. Which combinations of sharks resulted in the highest investments?

```python
multi_shark_deals["Shark_Combination"] = multi_shark_deals[[
    "AshneerGrover_deal", "AnupamMittal_deal", "AmanGupta_deal",
    "NamitaThapar_deal", "VineetaSingh_deal", "PeyushBansal_deal",
    "GhazalAlagh_deal"
]].astype(str).agg(' - '.join, axis=1)
shark_names = {
    "AshneerGrover_deal": "Ashneer Grover",
    "AnupamMittal_deal": "Anupam Mittal",
    "AmanGupta_deal": "Aman Gupta",
    "NamitaThapar_deal": "Namita Thapar",
    "VineetaSingh_deal": "Vineeta Singh",
    "PeyushBansal_deal": "Peyush Bansal",
    "GhazalAlagh_deal": "Ghazal Alagh"
}

def replace_sharks(row):
    sharks = []
    shark_columns = [
        "AshneerGrover_deal", "AnupamMittal_deal", "AmanGupta_deal",
        "NamitaThapar_deal", "VineetaSingh_deal", "PeyushBansal_deal",
        "GhazalAlagh_deal"
    ]
    for idx, shark_column in enumerate(shark_columns):
        if row[shark_column] == 1:
            sharks.append(shark_names[shark_column]) 
    return '-'.join(sharks)
multi_shark_deals["Shark_Combination_Names"] = multi_shark_deals.apply(replace_sharks, axis=1)
highest_investments_by_combo = multi_shark_deals.groupby("Shark_Combination_Names")["Deal_Investment_Amount"].sum().sort_values(ascending=False)
print("Top Shark Combinations with Highest Investments:")
print(highest_investments_by_combo.head())
```

```python
# Create a column for shark combinations as strings of '0' and '1'
multi_shark_deals["Shark_Combination"] = multi_shark_deals[[
    "AshneerGrover_deal", "AnupamMittal_deal", "AmanGupta_deal",
    "NamitaThapar_deal", "VineetaSingh_deal", "PeyushBansal_deal",
    "GhazalAlagh_deal"
]].astype(str).agg('-'.join, axis=1)

# Map 1 to shark name and 0 to empty string
shark_names = {
    "AshneerGrover_deal": "Ashneer Grover",
    "AnupamMittal_deal": "Anupam Mittal",
    "AmanGupta_deal": "Aman Gupta",
    "NamitaThapar_deal": "Namita Thapar",
    "VineetaSingh_deal": "Vineeta Singh",
    "PeyushBansal_deal": "Peyush Bansal",
    "GhazalAlagh_deal": "Ghazal Alagh"
}

# Replace 1 with the corresponding shark's name, and 0 with empty string
def replace_sharks(row):
    sharks = []
    shark_columns = [
        "AshneerGrover_deal", "AnupamMittal_deal", "AmanGupta_deal",
        "NamitaThapar_deal", "VineetaSingh_deal", "PeyushBansal_deal",
        "GhazalAlagh_deal"
    ]
    for idx, shark_column in enumerate(shark_columns):
        if row[shark_column] == 1:
            sharks.append(shark_names[shark_column])  # Add shark name if they invested
    return '-'.join(sharks)

# Apply the function to create the Shark_Combination_Names column
multi_shark_deals["Shark_Combination_Names"] = multi_shark_deals.apply(replace_sharks, axis=1)

# Group by Shark_Combination and sum up the Deal_Investment_Amount, then sort
highest_investments_by_combo = multi_shark_deals.groupby("Shark_Combination_Names")["Deal_Investment_Amount"].sum().sort_values(ascending=False)

# Display the top combinations with the highest investments
print("Top Shark Combinations with Highest Investments:")
print(highest_investments_by_combo.head())
```
